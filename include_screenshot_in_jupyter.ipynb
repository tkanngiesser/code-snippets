```python
from PIL import ImageGrab, Image
```

```python
from IPython.display import Image
```

```python
def clip(name=None, path='img', width=400, height=400):
    '''saves image from clipboard and displays saved image in jupyter'''
    im= ImageGrab.grabclipboard()
    im.save(f'{path}/{name}.png')
    return Image(filename=f'{path}/{name}.png', width=width, height=height)
```

```python
clip(name='hdi')
```
